```python
import csv
import pandas as pd
import math
import matplotlib.pyplot as plt

def read_csv_data():
    mean_data_list = []
    var_data_list = []
    horizon_plot_value = []
    for n in range(1,15):
        data = pd.read_csv(f'dat1/dat1.{n}.csv')
        var_data = data.var(axis=0)
        mean_data = data.mean(axis=0)
        mean_data_list.append(mean_data.item())
        var_data_list.append(var_data.item())
        horizon_plot_value.append(math.log10(2**n))
    
    return horizon_plot_value, mean_data_list, var_data_list
    
def plot_data(horizon_plot_value, mean_data_list, var_data_list):
    plt.plot(horizon_plot_value, mean_data_list, "o")
    plt.title('Mean Plot')
    plt.ylabel('Mean')
    plt.savefig('img/mean_plot.svg')
    plt.show()
    
    plt.title('Var Plot')
    plt.plot(horizon_plot_value, var_data_list, "o")
    plt.ylabel('Var')
    plt.savefig('img/var_plot.svg')
    plt.show()
    
x, y_mean, y_var =read_csv_data()
plot_data(x, y_mean, y_var)
```
