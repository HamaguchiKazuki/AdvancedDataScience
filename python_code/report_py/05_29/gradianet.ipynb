```python
def gradient(T, alpha_t):
    t = 0
    x = []
    x.insert(t, 3)
    f_x = []
    differential_f_x = []
    for _ in range(T):
        # print(f"{'#'*10} {t} {'#'*10}")
        f_x.insert(t, x[t] - np.log(x[t]))
        differential_f_x.insert(t, 1 - 1/x[t])
        # print(differential_f_x[t])
        x.insert(t+1, x[t] - alpha_t * differential_f_x[t])
        t += 1
        # print(x, len(x))
        # print(f_x, len(f_x))
    return x, f_x, differential_f_x 

def plot_gradient(x_t, f_x_t, T, alpha_t_label):
    t_line = np.linspace(0, T, T+1)
    fig, ax = plt.subplots(2)
    fig.subplots_adjust(wspace=0.4, hspace=0.6)
    ax[0].set(xlabel="t", ylabel="x_t", title="x=t, y=x_t Graph")
    ax[0].plot(t_line, x_t, "-", label=f"alpha_t={alpha_t_label}")
    ax[0].legend()
    ax[1].set(xlabel="t", ylabel="f(x_t)", title="x=t, y=f(x_t) Graph")
    ax[1].plot(t_line[:T], f_x_t, "-", label=f"alpha_t={alpha_t_label}")
    ax[1].legend()
    os.makedirs("img", exist_ok=True)
    fig.savefig(f'img/gradient_alpha_t={alpha_t_label}.png', format="png")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import os

t = 1
T = 100
alpha_t_list = [0.01, 2.0, 1.0/t, 1/np.sqrt(t)]
alpha_t_label_list = ["0.01", "2.0", "1.0t", "1np.sqrt(t)"]




for alpha_t, alpha_t_label in zip(alpha_t_list, alpha_t_label_list):
    x_t, f_x_t, _ = gradient(T, alpha_t)
    plot_gradient(x_t, f_x_t, T, alpha_t_label)
```
